```python
a= sns.load_dataset("penguins")

```

```python
a.columns
```

```python
# jointplot() multivariable view of complex data set
sns.jointplot(data=a, x="flipper_length_mm", y="bill_length_mm", hue="species")

# multiple kinds of plots to quickly give informative summaries of a dataset

```

```python
# it shows joint and marginal distributions for all pairwise relationships and for each variable, respectively

# pairplots


sns.pairplot(data=a, hue="species")
```

```python
sns.relplot(
    data=a,
    x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g"
)
```

```python
#  its functions have standardized parameters that can modify the semantic mappings for each plot

sns.set_theme(style="ticks", font_scale=1)
g = sns.relplot(data=a,x="bill_length_mm", y="bill_depth_mm", hue="body_mass_g",palette="crest", marker="x", s=100,)
g.set_axis_labels("Bill length (mm)", "Bill depth (mm)", labelpad=10)
g.legend.set_title("Body mass (g)")
g.figure.set_size_inches(6.5, 4.5)
g.ax.margins(.15)
g.despine(trim=True)
```
